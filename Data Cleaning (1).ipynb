# Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("heart_disease.csv.txt")
```

# null values

```python
df.isnull().sum()
```

# Representing null values in heatmap

```python

sns.heatmap(df.isnull())
```

```python

sns.countplot(x='AHD',data=df)
plt.show()
```

```python

sns.countplot(x='Ca',hue="Sex",data=df)
plt.show()
```

```python

sns.countplot(x='Thal',hue="Sex",data=df,palette='rocket_r')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x="Sex",y="Ca",data=df)
```

# Dropping null values column

```python
df.drop('Ca',axis=1,inplace=True)
df
```

# Filling null values

```python
m=df["Thal"].mode()[0]
m
```

```python
df["Thal"].fillna(m,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
#Heatmap represents cleaned data
```

```python
df["tt"]=pd.factorize(df.Thal)[0]
df
```

```python
sns.countplot(x="tt",hue="Sex",data=df,palette='YlGn')
```
